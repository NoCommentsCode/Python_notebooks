```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
theta_max = 100
sample = np.random.uniform(0, theta_max, N)

delta_1 = np.zeros(N - 1)
delta_2 = np.zeros(N - 1)
delta_3 = np.zeros(N - 1)
delta_4 = np.zeros(N - 1)
delta_5 = np.zeros(N - 1)

for n in range (1, N):
    _min = sample[:n].min()
    _max = sample[:n].max()
    delta_1[n - 1] = np.abs(theta_max - ((2 * np.sum(sample[0:n])) / n))
    delta_2[n - 1] = np.abs(theta_max - ((n + 1) * _min))
    delta_3[n - 1] = np.abs(theta_max - ((n + 1) / n) * _max)
    delta_4[n - 1] = np.abs(theta_max - (_max + _min))
    delta_5[n - 1] = np.abs(theta_max - (sample[0] + (1/(n)) * np.sum(sample[1:n])))
```

```python
plt.plot(delta_1, color='red')
plt.plot(delta_2, color='yellow')
plt.plot(delta_3, color='green')
plt.plot(delta_4, color='blue')
plt.plot(delta_5, color='black')
plt.grid()
plt.show()
```
