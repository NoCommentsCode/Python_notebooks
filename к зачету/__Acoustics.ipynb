```python
import numpy as np
from matplotlib import pyplot as plt
#%matplotlib inline
```

$\dfrac{du}{dt}=\dfrac{1}{ro}\dfrac{dro}{dt}$  
$\dfrac{dro}{dt}=ro*c^2\dfrac{dro}{dx}$

```python
rho_0 = 0.25
c_0 = 2
h = 0.05
Cu = 0.8
dt = Cu * h / c_0
```

```python
u_l, p_l = 1, 5 # x < 0
u_r, p_r = 0, 2 # x >= 0
```

```python
x = np.arange(-2, 2 + h, h)
```

```python
u = np.zeros_like(x)
p = np.zeros_like(x)
```

```python
left = np.where(x <= 0)
right = np.where(x > 0)
```

```python
u[left], p[left] = u_l, p_l
u[right], p[right] = u_r, p_r
```

```python
fig, axs = plt.subplots(2, 1, figsize = (10, 8))
axs[0].plot(x, u, label = 'u')
axs[1].plot(x, p, label = 'p')
axs[0].set_ylabel('u')
axs[1].set_ylabel('p')
for ax in axs:
    ax.set_xlabel('x')
    ax.legend()
    ax.grid(True)
```

```python
def get_BV(u_l, u_r, p_l, p_r):
    U = 0.5 * (u_l + u_r) - 0.5 * (p_r - p_l) / (rho_0 * c_0)
    P = 0.5 * (p_l + p_r) - 0.5 * (u_r - u_l) * (rho_0 * c_0)
    return U, P
```

```python
def godunov(u, p, nt, dt, dx, rho_0, c_0):
    u_n = np.zeros((nt, len(u)))
    p_n = np.zeros((nt, len(p)))
    u_n[0, :] = u.copy()
    p_n[0, :] = p.copy()
    for t in range(1, nt):
        U, P = get_BV(u[:-1], u[1:], p[:-1], p[1:])
        u_n[t, 1:-1] = u[1:-1] - dt / dx *(P[1:] - P[:-1]) / rho_0
        p_n[t, 1:-1] = p[1:-1] - dt / dx *(U[1:] - U[:-1]) * rho_0 * c_0 ** 2
        u_n[t, 0] = u[0]
        p_n[t, 0] = p[0]
        u_n[t, -1] = u[-1]
        p_n[t, -1] = p[-1]
        u = u_n[t].copy()
        p = p_n[t].copy()
    return u_n, p_n
```

```python
nt = 25
u_n, p_n = godunov(u, p, nt, dt, h, rho_0, c_0)
```

```python
fig, axs = plt.subplots(2, 1, figsize = (10, 8))
for step in (5, 10, 15, 20):
    axs[0].plot(x, u_n[step], '-o', label = step)
    axs[1].plot(x, p_n[step], '-o', label = step)
axs[0].set_ylabel('u')
axs[1].set_ylabel('p')
for ax in axs:
    ax.set_xlim(-1.5, 1.5)
    ax.set_xlabel('x')
    ax.legend()
    ax.grid(True)
```
