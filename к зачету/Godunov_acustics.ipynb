# Acoustics

Рассмотрим систему уравнений, которая описывает распространение плоских звуковых волн:

$$\frac{\partial u}{\partial t} + \frac{1}{\rho_0} \frac{\partial p}{\partial x} = 0,$$
$$\frac{\partial p}{\partial t} + {\rho_0 c_0^2} \frac{\partial u}{\partial x} = 0$$

Система _расщепилась_ на два уравнеия переноса

$$ \frac{\partial Y}{\partial t} + c_0  \frac{\partial Y}{\partial x} = 0,$$
$$\frac{\partial Z}{\partial t} - c_0  \frac{\partial Z}{\partial x} = 0,$$

где $$Y = u + \frac{p}{\rho_0 c_0},  Z = u - \frac{p}{\rho_0 c_0}$$

Решения уравнений: $$Y = f(x - c_0t), Z = g(x + c_0t)$$



Если вернуться к исходным переменым, получим
​
$$ u  = \frac{1}{2}[f(x - c_0t) + g(x+c_0t)],$$
$$ p  = \frac{\rho_0 c_0}{2}[f(x - c_0t) - g(x+c_0t)]$$

Если известны начальные условия $u(x,0) = u_0(x),\ p(x,0) = p_0(x)$, то функции $f$ и $g$ можно определить:

$$ u_0(x) = \frac{1}{2}[f(x) + g(x)],$$
$$ p_0(x)  = \frac{\rho_0 c_0}{2}[f(x ) + g(x)]$$

Отсюда 
$$f(x) = u_0(x) + \frac{p_0(x)}{\rho_0 c_0},\ g(x) = u_0(x) - \frac{p_0(x)}{\rho_0 c_0} $$

Окончательно, 

$$u(x,t) = \frac{u_0(x-c_0 t) + u_0(x+c_0t)}{2}  + \frac{p_0(x-c_0t) - p_0(x+c_0t)}{2\rho_0 c_0}$$
$$p(x,t) = \frac{p_0(x-c_0 t) + p_0(x+c_0t)}{2}  - \rho_0 c_0 \frac{u_0(x-c_0t) - u_0(x+c_0t)}{2} $$

Пусть в момент времени $t=0$ начальные условия имеют вид:

$$ u_0(x) = u_1,\, p_0(x) = p_1 \, при\ x < x^*,$$
$$ u_0(x) = u_{2},\, p_0(x) = p_{2} \, при\ x > x^*,$$

Тогда:
$$ \begin{cases}
u(x,0) = u_1, p(x,0) = p_1 & при\ x < x^* - c_0t \\ \\
u(x,0) = u_2, p(x,0) = p_2 & при\ x > x^* + c_0t \\ \\
u(x,0) = \frac{u_1 + u_2}{2}  + \frac{p_1 - p_2}{2\rho_0 c_0}\\
&при \; x^* - c_0t < x < x^* + c_0t \\
p(x,0) = \frac{p_1 + p_2}{2}  - \rho_0 c_0 \frac{u_1 - u_2}{2}  \\
\end{cases}$$

$$u^n = u_n - \frac{dt}{dx}\frac{1}{\rho_0} (P_{n + 0.5} - P_{n -0.5}) \\
p^n = p_n - \frac{dt}{dx}\rho_0 c_0^2 (U_{n + 0.5} - U_{n -0.5})$$

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 12
```

```python
rho_0 = 0.25
c_0 = 2
dx = 0.05
nt = 25
Cu = 0.9
dt = Cu * dx / c_0
```

```python
u_l, p_l = 1, 5 # x <= 0
u_r, p_r = 0, 2 # x > 0
```

```python
x = np.arange(-2,2 + dx, dx)
u = np.zeros_like(x)
p = np.zeros_like(x)

left = np.where(x<=0)
right = np.where(x>0)

u[left], p[left] = u_l, p_l
u[right], p[right] = u_r, p_r
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20,5))

axs[0].plot(x,u, label = 'u')
axs[1].plot(x,p, label = 'p')
axs[0].set_ylabel('u')
axs[1].set_ylabel('p')
for ax in axs:
    ax.set_xlabel('x')
    ax.legend()

    ax.grid(True)
```

```python
def Big_Values(u_l, u_r, p_l, p_r):
    U = 0.5 * (u_l + u_r) - 0.5 * (p_r - p_l) / (rho_0 * c_0)
    P = 0.5 * (p_l + p_r) - 0.5 * (u_r - u_l) * (rho_0 * c_0)
    return U, P
```

```python
def godunov(u, p, nt, dt, dx, rho_0, c_0):
    
    u_n = np.zeros((nt,len(u)))
    p_n = np.zeros((nt,len(p)))
    
    u_n[0,:] = u.copy()
    p_n[0,:] = p.copy()
    
    for t in range(1,nt):
        
        U, P = Big_Values(u[:-1],u[1:],p[:-1],p[1:])
        
        u_n[t, 1:-1] = u[1:-1] - dt / dx * (P[1:] - P[:-1]) / rho_0
        
        p_n[t,1:-1] = p[1:-1] - dt / dx * rho_0 * c_0**2 * (U[1:] - U[:-1])
        
        u_n[t,0] = u[0]
        p_n[t,0] = p[0]
        
        u_n[t,-1] = u[-1]
        p_n[t,-1] = p[-1]
        
        u = u_n[t].copy()
        p = p_n[t].copy()

    return u_n, p_n              
```

```python
u_n, p_n = godunov(u, p, nt, dt, dx, rho_0, c_0)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))
for step in (5,10,15,20):
    axs[0].plot(x, u_n[step],'-', label = step)
    axs[1].plot(x, p_n[step],'-', label = step)

axs[0].set_ylabel('u')
axs[1].set_ylabel('p')
for ax in axs:
    ax.set_xlim(-1,1)
    ax.set_xlabel('x')
    ax.legend()

    ax.grid(True)
```

$$ \frac{\partial{\textbf{U}}}{\partial{t}} + \textbf{A} \frac{\partial{\textbf{U}}}{\partial{x}} = 0 \\
\textbf{F} = \textbf{A} \textbf{U}$$

$$\textbf{A} = \begin{pmatrix}
0 & \frac{1}{\rho_0} \\
\rho_0 c_0^2 & 0 \\
\end{pmatrix}$$

Собственные числа матрицы $\textbf{A}$: $\lambda_1 = c_0$, $\lambda_2 = -c_0$

$$
\Lambda = 
\begin{pmatrix}
c_0 & 0 \\
0 & -c_0 \\
\end{pmatrix}
$$

$\Lambda^+ = 
\begin{pmatrix}
c_0 & 0 \\
0 & 0 \\
\end{pmatrix}$
$\Lambda^- = 
\begin{pmatrix}
0 & 0 \\
0 & -c_0 \\
\end{pmatrix}
$


Собственные вектора: 

$\lambda_1 = c_0 \rightarrow \begin{pmatrix} 1 \\ \rho_0c_0 \end{pmatrix} \;\;\;
\lambda_2 = -c_0 \rightarrow \begin{pmatrix} -1 \\ \rho_0c_0 \end{pmatrix} \;\;\;\;\;\;
\Rightarrow \;\;\;\;\;\;
K = \begin{pmatrix}
1 & -1 \\
\rho_0c_0 & \rho_0c_0 \\
\end{pmatrix}$

$$
K^{-1} = \frac{1}{2\rho_0c_0^2}
\begin{pmatrix}
\rho_0c_0^2 & c_0 \\
-\rho_0c_0^2 & c_0 \\
\end{pmatrix}
$$

$$\textbf{F}_{i + \frac{1}{2}} = \frac{1}{2}(\textbf{F}_i^n + \textbf{F}_{i+1}^n) - \frac{1}{2} |\textbf{A}| (\textbf{U}_i^n + \textbf{U}_{i+1}^n)$$

```python
U = np.array([u, p])
```

```python
A = np.array([[0, 1 / rho_0], [rho_0 * c_0**2, 0]])

mod_A = np.array([[c_0,0],[0,c_0]])
```

```python
def godunov_m(U, nt, dt, dx, rho_0, c_0):
    
    u_n = np.zeros((nt,len(u)))
    p_n = np.zeros((nt,len(p)))
    U_n = np.array([u_n, p_n])
    
    U_n[:, 0, :] = U.copy()
    
    
    for t in range(1,nt):
        F = 0.5 * (A @ U[:,:-1] + A @ U[:, 1:]) - 0.5 * mod_A @ (U[:,1:]-U[:,:-1])
        
        U_n[:,t,1:-1] = U[:,1:-1] - dt / dx * (F[:,1:] - F[:,:-1])
        
        U_n[:,t,0] = U[:,0]
        U_n[:,t,-1] = U[:,-1]
        
        U = U_n[:,t,:].copy()
        
    return U_n
```

```python
U_n = godunov_m(U, nt, dt, dx, rho_0, c_0)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10,8))
for step in (5,10,15,20):
    axs[0].plot(x, U_n[0, step],'-', label = step)
    axs[1].plot(x, U_n[1, step],'-', label = step)

axs[0].set_ylabel('u')
axs[1].set_ylabel('p')
for ax in axs:
    ax.set_xlim(-1,1)
    ax.set_xlabel('x')
    ax.legend()

    ax.grid(True)
```
