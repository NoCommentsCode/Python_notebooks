```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
from matplotlib import animation
from IPython.display import HTML
```

```python
nx = 81
nt = 101
rho_max = 10
u_max = 5
```

```python
def computeF(u):
    return u**2 / 2
```

```python
def flow(u_l, u_r):
    u_wl, u_wr = np.asarray([u_l, u_r])
    s = 0.5 * (u_wl + u_wr) #скорость скачска
    # |\
    if u_wl > u_wr: #скачек уплотнения
        u_star = u_l if s > 0 else u_r
    else: # u_wr > u_wl
        if u_wl > 0: # // волна разряжения
            u_star = u_l
        if u_wr < 0: # \\
            u_star = u_r   
        if u_wl < 0 and u_wr > 0: # \/
            u_star = u_max / 2
    F = computeF(u_star)
    return F
```

```python
def initial(nx):
    x = np.linspace(0, 4, nx - 1)
    u = np.exp(-(x - 1)**2 / 0.1)
    return u
```

```python
def godunov(nt, nx):
    Cu = 4
    dx = 4 / (nx - 1)
    dt = Cu * dx / u_max
    u = np.zeros((nt, nx - 1))
    u[0, :] = initial(nx=nx) # НУ
    u[:, 0] = u[0, 0]   # ГУ
    u[:, -1] = u[0, -1] # ГУ
    for t in range(nt-1):
        F = np.zeros(nx - 2)
        for i in range(nx - 2):
            F[i] = flow(u[t, i], u[t, i + 1])
        u[t + 1, 1:-1] = u[t, 1:-1] - dt/dx*(F[1:] - F[:-1])
    return u
```

```python
plt.figure(figsize=(10,4))
plt.grid()
plt.plot(np.linspace(0,4, nx-1), initial(nx=nx))
plt.ylabel('u')
plt.xlabel('x')
plt.title('Initial conditions');
```

```python
u = godunov(nt=nt, nx=nx)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(0,4),ylim=(-0.5,2),xlabel=('Distance'),ylabel=('Traffic density'))
line, = ax.plot([],[],color='#003366', lw=2);
def animate(data):
    x = np.linspace(0, 4, nx-1)
    y = data
    line.set_data(x,y)
    return line,
anim = animation.FuncAnimation(fig, animate, frames=u, interval=50)
HTML(anim.to_jshtml())
```
