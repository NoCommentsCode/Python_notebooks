## Lax - Friedrichs

$$\rho_i^{n+1} = \frac{1}{2}(\rho^n_{i+1}+\rho^n_{i-1}) - \frac{\Delta t}{2 \Delta x}(F^n_{i+1}-F^n_{i-1})$$

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16

from matplotlib import animation
from IPython.display import HTML
```

```python
nx = 81
nt = 101
rho_max = 10
u_max = 5
x = np.linspace(0, 2, nx - 1)
```

```python
def initial(rho_max, u_max, nx):
    dx = 4 / (nx - 1)
    x = np.linspace(0, 4, nx - 1)
    rho = np.ones_like(x) * 5
    rho[x >= 3] = rho_max
    return rho
```

```python
plt.figure(figsize=(10,4))
plt.plot(x, initial(rho_max=rho_max, u_max=u_max, nx=nx), color='#003366', ls='-', lw=3)
plt.ylabel('Traffic density')
plt.xlabel('Distance')
plt.ylim(-0.5,11.)
plt.title('Initial conditions');
```

```python
def computeF(u_max, rho_max, rho):
    return u_max * rho * (1 - rho / rho_max)
```

```python
def LaxFr(nt, rho_max, u_max, nx):
    Cu = 1
    dx = 4 / (nx - 1)
    dt = Cu * dx / u_max
    rho = np.zeros((nt, nx - 1))
    rho[0, :] = initial(rho_max=rho_max, u_max=u_max, nx=nx) # НУ
    rho[:, 0] = rho[0, 0]   # ГУ
    rho[:, -1] = rho[0, -1] # ГУ
    for t in range(nt - 1):
        F = computeF(u_max, rho_max, rho[t, :])
        rho[t + 1, 1:-1] = 0.5 * (rho[t, 2:] + rho[t, :-2]) - dt / (2*dx)*(F[2:]-F[:-2])        
    return rho
```

```python
rho = LaxFr(rho_max=rho_max, u_max=u_max, nx=nx, nt=nt)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(0,4),ylim=(-.5,11.5),xlabel=('Distance'),ylabel=('Traffic density'))
line, = ax.plot([],[], '-o',color='#003366', lw=2);
def animate(data):
    x = np.linspace(0, 4, nx-1)
    y = data
    line.set_data(x,y)
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, frames=rho, interval=50)
HTML(anim.to_jshtml())
```

## Lax - Wendroff

$$\rho_i^{n+1} = \rho_i^n - \frac{\Delta t}{2 \Delta x} \left(F^n_{i+1}-F^n_{i-1}\right)+ \frac{(\Delta t)^2}{4(\Delta x)^2} \left[ (J^n_{i+1}+J^n_i)(F^n_{i+1}-F^n_i)-(J^n_i+J^n_{i-1})(F^n_i-F^n_{i-1})\right]
$$

```python
plt.figure(figsize=(10,4))
plt.plot(x, initial(rho_max=rho_max, u_max=u_max, nx=nx), color='#003366', ls='-', lw=3)
plt.ylabel('Traffic density')
plt.xlabel('Distance')
plt.ylim(-0.5,11.)
plt.title('Initial conditions');
```

```python
def Jacobian(u_max, rho_max, rho):
    return u_max * (1 - 2 * rho / rho_max)
```

```python
def LaxWdr(nt, nx, rho_max, u_max):
    Cu = 1
    dx = 4 / (nx - 1)
    dt = Cu * dx / u_max
    rho = np.zeros((nt, nx - 1))
    rho[0, :] = initial(rho_max=rho_max, u_max=u_max, nx=nx) # НУ
    rho[:, 0] = rho[0, 0]   # ГУ
    rho[:, -1] = rho[0, -1] # ГУ
    for t in range(nt - 1):
        F = computeF(u_max, rho_max, rho[t, :])
        J = Jacobian(u_max, rho_max, rho[t, :])
        rho[t + 1,1:-1] = rho[t, 1: -1] - dt / (2 * dx) * (F[2:] - F[:-2]) \
                           + dt**2 / (4 * dx**2) * ( (J[2:] + J[1:-1]) * (F[2:] - F[1:-1]) \
                           - (J[1:-1] + J[:-2]) * (F[1:-1] - F[:-2]) )
    return rho
```

```python
rho = LaxWdr(rho_max=rho_max, u_max=u_max, nx=nx, nt=nt)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(0,4),ylim=(-.5,11.5),xlabel=('Distance'),ylabel=('Traffic density'))
line, = ax.plot([],[], '-o',color='#003366', lw=2);
def animate(data):
    x = np.linspace(0, 4, nx-1)
    y = data
    line.set_data(x,y)
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, frames=rho, interval=50)
HTML(anim.to_jshtml())
```

## MacCormack

$$\rho^*_i = \rho^n_i - \frac{\Delta t}{\Delta x} (F^n_{i+1}-F^n_{i}) $$

$$\rho^{n+1}_i = \frac{1}{2} (\rho^n_i + \rho^*_i - \frac{\Delta t}{\Delta x} (F^*_i - F^{*}_{i-1}))$$

```python
plt.figure(figsize=(10,4))
plt.plot(x, initial(rho_max=rho_max, u_max=u_max, nx=nx), color='#003366', ls='-', lw=3)
plt.ylabel('Traffic density')
plt.xlabel('Distance')
plt.ylim(-0.5,11.)
plt.title('Initial conditions');
```

```python
def Mac(nt, nx, u_max, rho_max):
    Cu = 1
    dx = 4 / (nx - 1)
    dt = Cu * dx / u_max
    rho = np.zeros((nt, nx - 1))
    rho[0, :] = initial(rho_max=rho_max, u_max=u_max, nx=nx) # НУ
    rho[:, 0] = rho[0, 0]   # ГУ
    rho_star = rho[0,:]
    for t in range(nt - 1):
        F = computeF(u_max, rho_max, rho[t, :])
        rho_star[:-1] = rho[t, :-1] - dt/dx * (F[1:] - F[:-1])
        F_star = computeF(u_max, rho_max, rho_star)
        rho[t + 1, 1:] = 0.5 * (rho[t, 1:] + rho_star[1:] - dt/dx * (F_star[1:] - F_star[:-1]))
        
    return rho
```

```python
rho = Mac(rho_max=rho_max, u_max=u_max, nx=nx, nt=nt)
```

```python
fig = plt.figure();
ax = plt.axes(xlim=(0,4),ylim=(-.5,11.5),xlabel=('Distance'),ylabel=('Traffic density'))
line, = ax.plot([],[], '-o',color='#003366', lw=2);
def animate(data):
    x = np.linspace(0, 4, nx-1)
    y = data
    line.set_data(x,y)
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, frames=rho, interval=50)
HTML(anim.to_jshtml())
```
