
# Классическая факторизация матриц для рекомендаций

Самая основная матричная модель факторизации для рекомендательных систем моделирует рейтинг $\hat{r}$, который пользователь $u$ дал бы элементу $i$ следующим образом: $$\hat{r}_{ui} = p_u^T q_i \\ $$
где $p_u = (p_u^1, p_u^2, p_u^3, ..., p_u^k)^T$ вектор, связанный с пользователем, $q_i^T = (q_i^1, q_i^2, q_i^3, ..., q_i^k)^T$  вектор, связанный с фильмом. Размерность векторов - это ранг модели, а компоненты называются _факторами_.

Мы можем собрать рейтинги _пользователь-элемент_ в матрицу  $\hat{R} = (\hat{r}_{ui})$. Во-первых, соберем вектора пользователей в матрицу:

$$P = \begin{pmatrix} \dots \; p_0 \; \dots \\ \dots \; p_1 \; \dots  \\ \vdots \\  \dots p_{n_{users}} \dots \end{pmatrix}$$

и вектора элементов в матрицу:

$$q = \begin{pmatrix} \dots \; q_0 \; \dots \\ \dots \; q_1 \; \dots  \\ \vdots \\  \dots q_{n_{items}} \dots \end{pmatrix}$$

Теперь мы можем выразить нашу матрицу $$\hat{R} = P Q^T.$$

Конечно, нет причин, по которым «истинная» матрица рейтенгов _пользователь-элемент_ $R = (r_{ui})$ должна иметь ранг $k$, и действительно, мы даже не узнаем эту «истинную» матрицу. Модель предполагает, что ее можно ___аппроксимировать___ факторизацией ранга $k$: $$R \approx \hat{R} = P Q^T.$$

Другими словами, мы хотим найти матрицу $P (n_{users} \times k)$ и матрицу $Q (n_{items} \times k)$, такую, что $\hat{R} = P Q^T$ аппроксимирует $R$.

# Смещение оценок пользователей и фильмов

На раннем этапе было осознано, что даже для коллаборативных фильтров рекомендательные системы работают намного лучше, если учитывать предвзятость пользователей и элементов.

Например, мы пытаемся предсказать рейтинги фильмов и знаем, что:

 - Боб ворчлив и оценивает фильмы в среднем на 2 звезды.
 - Алиса бодрая и веселая и оценивает вещи в среднем на 4 звезды.
Ясно, что 3 звезды от Боба сильно отличаются от 3-х звезд от Алисы. Если мы отслеживаем средний рейтинг фильма для каждого пользователя (предвзятость пользователя) и просто пытаемся предсказать разницу от этого среднего, система рекомендаций будет работать намного лучше.

Аналогично предположим:

"Змеиный полёт" получает средний рейтинг в 1 звезду (по всем пользователям)."Побег из Шоушенка" получает средний рейтинг 5 звезд. Тогда 3-звездочный рейтинг для "Змеиного полёта" сильно отличается от такой же оценки "Побега из Шоушенка". Если мы отслеживаем средний пользовательский рейтинг для каждого фильма (смещение элемента) и просто пытаемся предсказать разницу от этого среднего, система рекомендаций также будет работать намного лучше.

Для модели явных оценок, один из способов учесть предвзятость пользователя - смоделировать фактические рейтинги пользователь-элемент как $$r_{ui} \approx b_u + p^T_u q_i,$$
 
где $b_u$ - предвзятость пользователя $u$. 

Аналогичным образом мы можем учесть смещение по элементам:

$$r_{ui} \approx b_i + p^T_u q_i,$$ 

здесь $b_i$ - смещение для предмета $i$.

Окончательно, учтем оба смещения:

$$r_{ui} \approx b_i + b_u + p^T_u q_i.$$

Чтобы найти неизвестные оценки, мы минимизируем следующую регуляризованную функцию потерь:

$$ L = \sum_{u, i \in observed\;ratings} \left(r_{ui} - \hat{r}_{ui} \right)^2 +
\lambda\left(b_i^2 + b_u^2 + ||q_i||^2 + ||p_u||^2\right)$$

Минимизация выполняется очень простым стохастическим градиентным спуском:

$$b_u \leftarrow b_u + \gamma (e_{ui} - \lambda b_u)\\
b_i \leftarrow b_i + \gamma (e_{ui} - \lambda b_i)\\
p_u \leftarrow p_u + \gamma (e_{ui} \cdot q_i - \lambda p_u)\\
q_i \leftarrow q_i + \gamma (e_{ui} \cdot p_u - \lambda q_i)$$

где $e_{ui} = r_{ui} - \hat{r}_{ui}$. Эти шаги выполняются по всем рейтингам набора и повторяются n_epochs раз. Базовые показатели инициализируются нулями. Факторы пользователей и элементов инициализируются случайным образом в соответствии с нормальным распределением, которое можно настроить с помощью параметра _init_std_dev_. По умолчанию он 0.1.

Вы также можете контролировать скорость обучения $\gamma$ и член регуляризации $\lambda$. Оба могут быть разными для каждого типа параметра. По умолчанию они 0.005 и 0.02 соответственно.

В данной работе реализованы оба алгоритма (со смещением и без) + генетический алгоритм оптимизации.

Также стоит заметить, что предсказание оценок не наша конечная цель. По завершении работы алгоритм должен решить рекомендовать фильм пользователю или нет. Следовательно, нам не стоит сильно фокусироваться на предсказании низких оценок: не важно, что мы предскажем 1 или 2 - этот фильм мы точно не будем рекомендовать.

Итак, как мы же будем оценивать качество нашего алгоритма?
Для этого мы уже имеем несколько критериев.
- Функция потерь
- Диапазон предсказанных значений 
- Суммарная ошибка (на отложенной выборке)
- Корень среднего квадрата ошибки (на отложенной выборке)
- Точность предсказания каждой из оценок (на отложенной выборке)

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python


class SVD():
    def __init__(self, data):
        """
        data - pd.DataFrame (n_users, n_items)
        """
        self.data = data
        self.X = data.to_numpy()  # матрица известных оценок (целые числа)
        (self.U, self.I) = np.where(self.X > 0.5) # пары (пользователь, фильм) в которых есть оценка
        self.A = None # итоговая матрица оценок (заранее известных и предсказанных) 
        self.q = None # матрица латентных признаков для фильмов
        self.p = None # матрица латентных признаков для зрителей
        self.bu = None # вектор предвзятостей пользователей
        self.bi = None # вектор смещения фильмов
        self.m = None # общее смещение оценок #глобальное смещение
        self.e = None # матрица ошибок при оценивании фильмов
        self.error = None # вектор суммарной абсолютной ошибки предсказания на каждой итерации


    # в нашем исследовании факторизация матрицы производится двумя способами: генетическим алгоритмом и методом SGD
    # использование генетического алгоритма
    def fit_Genetic(self, k=4, N_base_population=100, l2_reg=0.5, border_coef=1, ratio_from_svd=0.75,
                    num_cross=5, num_mut=5, num_worst=1, max_iteration=100, max_permanent_iteration=10, 
                    var=0.5, loc_glob=1, var_glob=0.5, verbose=False, Fine_bez=1e10, Fine_a5=1e4):
      
        gen_alg = GeneticAlgo(base_matrix=self.X, k=k, N_base_population=N_base_population, l2_reg=l2_reg, border_coef=border_coef, 
                              ratio_from_svd=ratio_from_svd, num_cross=num_cross, num_mut=num_mut, num_worst=num_worst, var=var, 
                              loc_glob=loc_glob, var_glob=var_glob, verbose=verbose, Fine_bez=Fine_bez, Fine_a5=Fine_a5)
        
        gen_alg.optimisation_search(MaxIteration=max_iteration, MaxPermanentAnswer=max_permanent_iteration)

        self.global_bias = gen_alg.BGLOB
        self.q = gen_alg.Q
        self.p = gen_alg.P
        self.A = self.p @ self.q.T + self.global_bias

    # использование метода SGD
    def fit(self, k = 4, gamma = 0.005, l = 0.02, n_epochs=20, init_std_dev=0.1, verbose=False, level=100, bias=True):
        self.A = np.zeros_like(self.X)
        n_users = self.X.shape[0]
        n_items = self.X.shape[1]
        self.q = np.random.normal(scale=init_std_dev, size=(n_items, k))
        self.p = np.random.normal(scale=init_std_dev, size=(n_users, k))
        self.m = np.mean(self.X[self.U,self.I])
        self.e = np.zeros_like(self.A)
        self.error = np.zeros(n_epochs)


        def step_bias(u, i):
            self.bu[u] = self.bu[u] + gamma * (self.e[u,i] - l * self.bu[u])
            self.bi[i] = self.bi[i] + gamma * (self.e[u,i] - l * self.bi[i])
            self.p[u] = self.p[u] + gamma * (self.e[u,i] * self.q[i] - l * self.p[u])
            self.q[i] = self.q[i] + gamma * (self.e[u,i] * self.p[u] - l * self.q[i])

        def step(u, i):
            self.p[u] = self.p[u] + gamma * (self.e[u,i] * self.q[i] - l * self.p[u])
            self.q[i] = self.q[i] + gamma * (self.e[u,i] * self.p[u] - l * self.q[i])

        if bias:
          self.m = 0
          self.bu = np.zeros(n_users)
          self.bi = np.zeros(n_items)
          for iter in range(n_epochs):
              order = np.random.permutation(len(self.U))
              for (u,i) in zip(self.U[order], self.I[order]):
                  self.A[u, i] = self.q[i].T @ self.p[u] + self.bu[u] + self.bi[i] + self.m
                  self.e[u,i] = self.X[u,i] - self.A[u, i]
                  step_bias(u,i)
              self.error[iter] = np.sum(abs(self.e))
              if verbose: print(f'[{iter}] : MAE = ', self.error[iter])
              if (iter%level == 0): gamma = gamma/2
          self.A = self.m + self.bu.reshape(n_users,1) + self.bi.reshape(1,n_items) + self.p @ self.q.T 

          
        else:
          X_st = self.X - self.m
          P, S, Q = np.linalg.svd(X_st, full_matrices=False)
          P = P @ np.diag(S**0.5)
          Q = np.diag(S**0.5) @ Q
          self.p = P[:, :k]
          self.q = Q.T[:, :k]
          for iter in range(n_epochs):
              order = np.random.permutation(len(self.U))
              for (u,i) in zip(self.U[order], self.I[order]):
                  self.A[u, i] = self.q[i].T @ self.p[u]
                  self.e[u,i] = X_st[u,i] - self.A[u, i]
                  step(u,i)
              self.error[iter] = np.sum(abs(self.e))
              if verbose: print(f'[{iter}] : MAE = ', self.error[iter])
              if (iter%level == 0): gamma = gamma/2
          self.A = self.m + self.p @ self.q.T        
        
    # Функция для оценки предсказания
    # Выводит следующую информацию:
    # - Все уникальные значения предсказания
    # - Суммарная ошибка предсказания
    # - Максимальную абсолютную ошибку предсказания 
    # - Корень среднего квадарата ошибки
    # А также считает точность предсказания для каждой оценки

    def estimate(self, plot=True):
        assert os.path.exists('Matrix_test.csv'), "Добавьте 'Matrix_test.csv' в директорию"
        (Z_r, Z_c) = np.where(self.X < 0.5)
        test = pd.read_csv('Matrix_test.csv', index_col=0).to_numpy()
        X_pred = np.round(self.A, 0)
        test_flat = test[Z_r, Z_c]
        pred_flat = X_pred[Z_r, Z_c]
        error = pred_flat - test_flat
        accuracy = dict()
        for mark in (1, 2, 3, 4, 5):
          inds = np.where(abs(test_flat - mark) < 0.5)[0]
          accuracy[mark] = np.round(np.sum(abs(error[inds]) < 0.5) / len(inds),2)
        self.accuracy = accuracy
        print('Все уникальные значения предсказания: ', np.unique(X_pred))
        print('Суммарная ошибка предсказания: ', np.sum(abs(error)))
        print('max_absolute_error: ', np.max(np.abs(error)))
        print('RMSE: ', np.sqrt(np.mean(error**2)))
        self.RMSE = np.sqrt(np.mean(error**2))
        if plot:
            fig, axs =plt.subplots(1,3, figsize=(24,8))
            fig.suptitle('Распределение', fontsize = 30)
            axs[0].set_title('Предсказанных оценок')
            sns.countplot(x = X_pred[Z_r, Z_c], ax=axs[0])
            axs[1].set_title('Реальных оценок')
            sns.countplot(x = test[Z_r, Z_c], ax=axs[1])
            axs[2].set_title('Ошибок')
            sns.countplot(x = error, ax=axs[2])
        
    def error_plot(self):
        plt.plot(np.arange(len(self.error)), self.error)
        
    def factors(self):
        p = pd.DataFrame(self.p, index=data.index)
        p.index.name = 'userId'
        
        q = pd.DataFrame(self.q, index=data.columns)
        q.index.name = 'movieId'
        
        fig, axs = plt.subplots(1, 2, figsize=(20,18))
        fig.suptitle('Latent Space', size=30, va='bottom')
        axs[0].set_title('Users')
        axs[1].set_title('Movies')

        sns.heatmap(ax=axs[0], data=p.head(30), annot=True, center= 0)
        sns.heatmap(ax=axs[1], data=q, annot=True, center= 0)
        
```

```python
def sparse_dist(base_matrix, predict, U, I):
    """
    Вычисление расстояния между матрицами по выделенным ячейкам матрицы
    
    Вход:
        base_matrix - np.array(n_users, n_items) - разреженная матрица оценок
        predict - np.array(n_users, n_items) - матрица предсказанных оценок
        U - np.array(n_ind) - массив номеров строчек, по которым считается расстояние
        I - np.array(n_ind) - массив номеров столбцов, по которым считается расстояние
        
    Выход:
        sqr_sum - float - сумма квадратов разностей значений по выделенным ячейкам
    """
    errors = base_matrix[U,I] - predict[U,I]
    return np.sum(errors**2)



class species():
    """
    Класс-особь генетического алгоритма.
    Состоит из матрицы латентных признаков для зрителей p и матрицы латентных признаков для фильмов q
    """
    def __init__(self, p , q, bu, bi, bglob):
        """
        p - np.array(n_users, k) - предсказанная матрица латентных признаков для зрителей
        q - np.array(n_items, k) - предсказанная матрица латентных признаков для фильмов
        """
        self.p = p
        self.q = q
        self.bu = bu
        self.bi = bi
        self.bglob = bglob

        
def Factorization_loss_reg(base_matrix, U, I, p, q, bu, bi, bglob, reg=1.0, Fine_bez=1e10, Fine_a5=1e4):
    """
    Вычисление функции потерь для Факторизации матрицы
    
    Вход:
        base_matrix - np.array(n_users, n_items) - разреженная матрица оценок 
                                                   (возможно, в ней уже учтено смещение)
        U - np.array(n_ind) - массив номеров строчек, по которым считается расстояние
        I - np.array(n_ind) - массив номеров столбцов, по которым считается расстояние
        p - np.array(n_users, k) - предсказанная матрица латентных признаков для зрителей
        q - np.array(n_items, k) - предсказанная матрица латентных признаков для фильмов
        k - int - число латентных признаков
        reg - float - коэффициент регуляризации
    
    Выход:
        L - float - значение функции потерь
    """
    n_users, n_items = base_matrix.shape
    
  
    predict = bu.reshape(n_users,1) + bi.reshape(1,n_items) + p @ q.T + bglob
    
    bez = np.sum(predict<=0.5) # bellow or equal zero
    a5 = np.sum(predict>=5.5) # above 5
    spar_d = sparse_dist(base_matrix=base_matrix, predict=predict, U=U, I=I)
    L = spar_d + reg*(np.sum(p[U]**2) + np.sum(q[I]**2)) + Fine_bez*bez + Fine_a5*a5
    
    
    return L        
        
        

class GeneticAlgo():
    """
    Генетический алгоритм для решения задачи оптимизации при факторизации матрицы.
    """
    
    def __init__(self, base_matrix, k, N_base_population=20, l2_reg=1.0, border_coef=2, ratio_from_svd=0.75, 
                 num_cross=5, num_mut=5, num_worst=1, var=0.5, loc_glob=1, var_glob=0.5, verbose=False, Fine_a5=1e4, Fine_bez=1e10):
        """
        base_matrix - np.array(n_user, n_items) - разреженная матрица оценок 
                                                  (возмонжно, в ней уже учтено смещение)
        k - int - число латентных признаков
        N_base_population - int - поддерживаемый размер популяции
        l2_reg - float - коэффициент l2-регуляризации
        border_coef - float - коэффициент, определяющий в каком диапазоне находятся значения латентных признаков
        ratio_from_svd - float - доля случаных строк, взятых из матриц P_SPARSE_cutted и Q_SPARSE_cutted для 
                                 инициализации особей
        num_cross - int - число скрещиваний на этапе кроссинга
        num_mut - int - число мутаций на этапе мутирования
        num_worst - int - число худжих особоей, сохраняемых на этапе отбора
        var - float - вариация 
        """
        self.base_matrix = base_matrix
        self.k = k
        self.N_base_population = N_base_population
        self.l2_reg = l2_reg
        self.ratio_from_svd = ratio_from_svd 
        self.num_cross = num_cross
        self.num_mut = num_mut
        self.num_worst = num_worst
        self.var = var
        self.loc_glob = loc_glob
        self.var_glob = var_glob
        self.verbose = verbose
        self.Fine_a5 = Fine_a5
        self.Fine_bez = Fine_bez
        
        (self.U, self.I) = np.where(base_matrix > 0.5) 
        self.N_current_population = 0 
        self.population = None       # массив всех особей популяции
        self.all_adaptivities = None # массив значений функции приспособленнсти для каждой особи
        
        u, s , vt = np.linalg.svd(self.base_matrix, full_matrices=False) # p ~ u, q ~ vt.T
        # матрицы латентных признаков для исходной разреженной матрицы
        self.P_SPARSE = u @ np.diag(np.sqrt(s)) # (n_users, K)
        self.Q_SPARSE = (np.diag(np.sqrt(s)) @ vt).T #(n_items, K)
        
        # обрезанные матрицы латентных признаков
        self.P_SPARSE_cutted = self.P_SPARSE[:,:self.k]
        self.Q_SPARSE_cutted = self.Q_SPARSE[:,:self.k]
        
        self.MaxAbsP = border_coef*np.max(np.abs(self.P_SPARSE_cutted))
        self.MaxAbsQ = border_coef*np.max(np.abs(self.Q_SPARSE_cutted))
        return
    
    
    def adaptivity_function(self, spec):
        """
        Вычисление функции приспособленности для особи
        
        Вход:
            spec - species(p,q) - особь
        
        Выход:
            adaptivity - float - значение функции приспособленности
        """
        adaptivity = Factorization_loss_reg(base_matrix=self.base_matrix, U=self.U, I=self.I,
                                            p=spec.p, q=spec.q, bu=spec.bu, bi=spec.bi, bglob=spec.bglob,
                                            reg=self.l2_reg, Fine_bez=self.Fine_bez, Fine_a5=self.Fine_a5)
        return adaptivity    
    

    def generate_species(self,):
        """
        Функция, генериующая особь в соответствии с количеством латентных признаков self.k.
        
        Выход:
            spec - species(p,q) - особь, матрицы которой имеют следующие размерности:
                                  p - (n_users, k)
                                  q - (n_items, k)
        """
        n_users, n_items = self.base_matrix.shape
#         p = np.random.uniform(low=-self.MaxAbsP, high=self.MaxAbsP, size=(n_users,self.k))
#         q = np.random.uniform(low=-self.MaxAbsQ, high=self.MaxAbsQ, size=(n_items,self.k))
        
#         u = np.unique(self.U)
#         u = np.random.choice(u, size=np.int64(np.round(n_users*self.ratio_from_svd)), replace=False)
    
#         i = np.unique(self.I)
#         i = np.random.choice(i, size=np.int64(np.round(n_items*self.ratio_from_svd)), replace=False)
        
        
#         p[u] = self.P_SPARSE_cutted[u]
#         q[i] = self.Q_SPARSE_cutted[i]
        
        Noise_BM = self.base_matrix + np.random.normal(loc=0.0, scale=self.var, size=self.base_matrix.shape)
        
        u, s , vt = np.linalg.svd(Noise_BM, full_matrices=False)
        P = u @ np.diag(np.sqrt(s))
        Q = (np.diag(np.sqrt(s)) @ vt).T
        
#         bu = np.empty(n_users)
#         bi = np.empty(n_items)
#         for i in range(n_users):
#             bu[i] = np.sum(self.base_matrix[i]) /np.sum(self.base_matrix[i]>0) - 3
#         for i in range(n_items):
#             bi[i] = np.sum(self.base_matrix[:,i]) / np.sum(self.base_matrix[:,i]>0) - 3

        bu = np.zeros(n_users)
        bi = np.zeros(n_items)
    
        bglob = np.random.normal(loc=self.loc_glob, scale=self.var_glob)
        
        p = P[:,:self.k]
        q = Q[:,:self.k]
        spec = species(p=p, q=q, bu=bu, bi=bi, bglob=bglob)
        return spec
    

    def sort_population(self,):
        """
        Сортировка списков self.population и self.all_adaptivities по возрастанию значения
        функции приспособленности.
        """
        ind = np.argsort(self.all_adaptivities)
        self.population = self.population[ind]
        self.all_adaptivities = self.all_adaptivities[ind]
        return    


    def generate_initial_population(self,):
        """
        Инициализация начальной популяции.
        
        Выход:
            заполнение вектора особей self.population и вектора значений их функции присполсобленности
        """
        self.population = np.empty(self.N_base_population, dtype=species)
        self.all_adaptivities = np.empty(self.N_base_population)
#         first_species = species(p=self.P_SPARSE_cutted, q=self.Q_SPARSE_cutted)
#         self.population[0] = first_species
#         self.all_adaptivities[0] = self.adaptivity_function(first_species)
        
        self.N_current_population = self.N_base_population
        for i in range(0,self.N_base_population):
            spec = self.generate_species()
            self.population[i] = spec
            self.all_adaptivities[i] = self.adaptivity_function(spec)
        
        self.sort_population()
        return
    

    def Kr(self, sp1, sp2, f1, f2):
        """
        Оператор кроссинга. Порождение двух новых особей на основе уже имеющихся двух.
        Точка скрещивания вычисляется пропорционально функции приспособленности: 
        каждая матрица каждой особи представляется 1-D numpy-массивом, 
        а затем в нужной точке эти соответсвующие массивы p и q от разных особей скрещиваются.
        
        Вход:
            sp1, sp2 - species - две исходные особи
            f1, f2 - float - значения функции приспособленности для каждой исхожной особи
            
        Выход:
            new_sp1, new_sp2 - species - две новые особи
        """
        n_users, n_items = self.base_matrix.shape
        p1, q1 = sp1.p.ravel(), sp1.q.ravel() # вытянули матрицы в 1-D массив
        p2, q2 = sp2.p.ravel(), sp2.q.ravel()
        
        Np, Nq = p1.shape[0], q1.shape[0] # сколько элементов в каждом типе матриц
        # вычисление точки скрещивания
        # proportion = f1 / (f1 + f2)  
#         Pcp = np.int64(np.round(Np * proportion)) # P_cross_point
#         Qcp = np.int64(np.round(Nq * proportion)) # Q_cross_point
    
        Pcp = np.random.randint(low=1, high=Np)
        Qcp = np.random.randint(low=1, high=Nq)
    
        # содание скрещенных особей
        new_p1, new_p2 = np.zeros_like(p1), np.zeros_like(p2)
        new_p1[:Pcp], new_p2[:Pcp] = p1[:Pcp], p2[:Pcp] # до точки кроссинга всё как было
        new_p1[Pcp:], new_p2[Pcp:] = p2[Pcp:], p1[Pcp:] # после точки кроссинга идёт смешение
        
        new_q1, new_q2 = np.zeros_like(q1), np.zeros_like(q2)
        new_q1[:Qcp], new_q2[:Qcp] = q1[:Qcp], q2[:Qcp] 
        new_q1[Qcp:], new_q2[Qcp:] = q2[Qcp:], q1[Qcp:] 
        
        new_P1 = new_p1.reshape((n_users, self.k))
        new_P2 = new_p2.reshape((n_users, self.k))
        new_Q1 = new_q1.reshape((n_items, self.k))
        new_Q2 = new_q2.reshape((n_items, self.k))
        
        new_sp1 = species(p=new_P1, q=new_Q1, bu=sp2.bu, bi=sp2.bi, bglob=sp2.bglob)
        new_sp2 = species(p=new_P2, q=new_Q2, bu=sp1.bu, bi=sp1.bi, bglob=sp1.bglob)
        return new_sp1, new_sp2
    
    
    def crossing(self,):
        """
        Осуществление этапа кроссинга: отбор и дальнейшее скрещивание.
        По итогу работы это функции исходная популяция увеличится на self.num_cross*2 особей, полученных
        путём скрещивания.
        """
        # чтобы выбрать лучших особей, их нужно отсортировать
        self.sort_population()
        
        new_species = np.empty(2*self.num_cross, dtype=species)
        new_adaptivities = np.empty(2*self.num_cross)
        for i in range(self.num_cross):
            # выберем две случайные особи для скрещивания 
            # i1 = np.random.randint(low=0, high=self.num_cross) 
            i1 = 0
            i2 = np.random.randint(low=1, high=self.num_cross)  
            while (i1 == i2):
                i2 = np.random.randint(low=0, high=self.num_cross)
            sp1, sp2 = self.Kr(self.population[i1], self.population[i2], 
                               self.all_adaptivities[i1], self.all_adaptivities[i2])
            
            new_species[2*i] = sp1
            new_adaptivities[2*i] = self.adaptivity_function(sp1)
            
            new_species[2*i+1]=sp2
            new_adaptivities[2*i+1]=self.adaptivity_function(sp2)
            
        # добавим новых особей в основную популяцию
        self.N_current_population = self.N_current_population + 2*self.num_cross
        self.population = np.concatenate((self.population, new_species))
        self.all_adaptivities = np.concatenate((self.all_adaptivities, new_adaptivities))
        return
    
    def mutation(self,):
        """
        Осуществление этапа мутации. Способ: добавление новой случайной особи в популяцию.
        По итогу работы этой функции в популяции появится self.num_mut новых мутировавших особей
        """
        new_species = np.empty(self.num_mut, dtype=species)
        new_adaptivities = np.empty(self.num_mut)
        for i in range(self.num_mut):
            n_users, n_items = self.base_matrix.shape
#             best_p, best_q = self.population[0].p, self.population[0].q
#             predict = best_p @ best_q.T
#             (U,I) = np.where(predict>0.5)
#             U = np.unique(U)
#             I = np.unique(I)
            
            # sp = self.generate_species()
#             sp.p[U] = best_p[U]
#             sp.q[I] = best_q[I]
            best_p, best_q, best_b = self.population[0].p, self.population[0].q, self.population[0].bglob
            best_p[np.random.randint(n_users)] += np.random.normal(loc=0.0, scale=self.var, size=self.k)
            best_q[np.random.randint(n_items)] += np.random.normal(loc=0.0, scale=self.var, size=self.k) 
            best_b += np.random.normal(loc=0.0, scale=self.var) 
            bu = np.zeros(n_users)
            bi = np.zeros(n_items)
            sp = species(p=best_p, q=best_q, bu=bu, bi=bi, bglob=best_b)
            
            new_species[i] = sp
            new_adaptivities[i] = self.adaptivity_function(sp)
        
        # добавим новых особей в основную популяцию
        self.N_current_population = self.N_current_population + self.num_mut
        self.population = np.concatenate((self.population, new_species))
        self.all_adaptivities = np.concatenate((self.all_adaptivities, new_adaptivities))
        return

    
    def selection(self,):
        """
        Осуществление этапа отбора. После него должно остаться N_base_population особей.
        Метод: взять self.num_worst худших особей и (self.N_base_population-self.num_worst) лучших
        """
        # чтобы выбрать худших и лучших особей, их нужно отсортировать
        self.sort_population()
        
        new_species = np.empty(self.N_base_population, dtype=species)
        new_adaptivities = np.empty(self.N_base_population)
        
        #записываем лучших в начало
        new_species[:self.N_base_population-self.num_worst] = self.population[:self.N_base_population-self.num_worst]
        new_adaptivities[:self.N_base_population-self.num_worst] = self.all_adaptivities[:self.N_base_population-self.num_worst]
        
        #записывем худших в конец
        new_species[self.N_base_population-self.num_worst:] = self.population[-self.num_worst:]
        new_adaptivities[self.N_base_population-self.num_worst:] = self.all_adaptivities[-self.num_worst:]
        
        #перезапишем основуную популяцию
        self.population = new_species
        self.all_adaptivities = new_adaptivities
        self.N_current_population = self.N_base_population
        return
    
    
    def optimisation_search(self, MaxIteration=10, MaxPermanentAnswer=10):
        """
        Решение оптимизационной задачи факторизации матрицы. По результату работы алгоритма создадуться 
        и заполнятся следующие поля:
            num_iteration - int - число итераций, которое потребовал оптимизационный процесс
            n_error - np.array(num_iteration, ) - вектор минимальных ошибок на каждой итерации
            P - np.array(n_users, k) - предсказанная матрица латентных признаков для зрителей
            Q - np.array(n_items, k) - предсказанная матрица латентных признаков для фильмов
        
        Вход:
            MaxIteration - int - максимальное количество совершённых циклов генетического отбора
            MaxPermanentAnswer - int - критерий остановки процесса оптимизации. Если MaxPermanentAnswer раз
                                       подряд алгоритм даёт одну и ту же ошибку, то процесс оптимизации заканчивается
        """
        n_error = np.empty(MaxIteration)
        
        self.generate_initial_population()
        n_error[0] = self.all_adaptivities[0]
        if self.verbose:
          print("initial: ", n_error[0])
        
        i_ans = 0 #счётчик, считающий количество одинаковых ответов подряд
        for i in range(1, MaxIteration):
            if self.verbose:
                print(f'[{i}]', end='')
            self.crossing()
            self.mutation()
            self.selection()
            
            n_error[i] = self.all_adaptivities[0]
            if n_error[i] == n_error[i-1]:
                i_ans += 1
                if i_ans==MaxPermanentAnswer:
                    n_error = n_error[:i+1]
                    break
            else:
                i_ans = 0
                if self.verbose:
                    print(f" : {n_error[i]}")
            if self.verbose:
              print('\r', end='')
            
        # после работы этого цикла ответ на задачу содержится в самой первой особи популяции
        self.num_iteration = i
        self.BU = self.population[0].bu 
        self.BI = self.population[0].bi
        self.BGLOB =  self.population[0].bglob
        self.P = self.population[0].p
        self.Q = self.population[0].q
        return
```
