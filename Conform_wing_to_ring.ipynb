```python
import numpy as np
from matplotlib import pyplot as plt
import cmath as cm
```

```python

file = open("prof.dat", 'r')
text = file.readlines()
size = len(text)
x = np.zeros(size)
y = np.zeros(size)
i = 0
for line in text:
    l = line[:-1].split(' ')
    x[i] = float(l[0])
    y[i] = float(l[1])
    i += 1
"""
data = np.loadtxt("frof.dat")
print(data)
"""
```

```python
plt.plot(x, y);
```

```python
Z = np.zeros(size, dtype=complex)
Z = x + 1j * y
```

```python
zt = 1
zn = 0.009
t = 0.2772

A = zt
B = zn
k = 2 - t / cm.pi
a = (zn - zt) / (2. * k)
b = -a
```

```python
W = (Z - A) / (Z - B)
V = W ** (1. / k)
Z1 = (a - b * V) / (1. - V)
```

```python
center = np.sum(Z1) / size
print(center)
```

```python
x1 = np.array([i.real for i in Z1])
y1 = np.array([i.imag for i in Z1])
```

```python
plt.figure(figsize=(5,5))
#plt.xlim([-0.35, 0.35])
#plt.ylim([-0.3, 0.3])
plt.plot(x1, y1);
```
