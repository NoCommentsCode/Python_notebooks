# Одномерная функция растяжения / сжатия

```python
def show_grid(points):
    plt.figure(figsize=(20,5))
    plt.scatter(points, np.ones_like(points), c='black', s=50)
    plt.xticks(points)
    plt.grid(True)
```

```python
# расчётная область
x_A, x_E = -5, 5
N = 20
x = np.linspace(x_A, x_E, N)

show_grid(x)
```

Нормализация координат:  
$$\displaystyle \eta^* = \dfrac{\eta - \eta_A}{\eta_E - \eta_A}; \;\; \eta^* \in [0;1], \;\; \eta_A \le \eta \le \eta_E$$

```python
# перевод в нормализованный вид
def normalization(points):
    pA, pE = points[0], points[-1]
    points_star = (points-pA)/(pE - pA)
    return points_star

def denormalization(points_star, pA, pE):
    return points_star*(pE-pA) + pA

x_star = normalization(x)
show_grid(x_star)
show_grid(denormalization(x_star, x_A, x_E))
```

Одномерная функция растяжения Робертса:

$$S = P \eta^* + (1-P)\Bigl(1-\dfrac{\tanh{[Q(1-\eta^*)]}}{\tanh{Q}}\Bigr),$$  
где $P$ и $Q$ - параметры, определяющие сгущение:
* чем ближе $P$ к $1$ тем больше сгущение к точке А
* $Q$ определяет отклонение от линейного изменения

```python
def oneD_compress_function(x_star, P=1, Q=1e-10):
    S = P*x_star + (1-P)*(1 - np.tanh(Q*(1 - x_star))/np.tanh(Q))
    return S

S = oneD_compress_function(x_star, P=0.1, Q=2)
show_grid(S)
```

```python
params = {'P' : np.linspace(0,10,10),
          'Q' : np.linspace(1, 5, 5)}
for p in params['P']:
    for q in params['Q']:
        S = oneD_compress_function(x_star, P=p, Q=q)
        show_grid(S)
        plt.title(f"P = {p}, Q = {q}", size=16)
```

# Задача с двумерным каналом

```python
def f1(x):
    return np.sqrt(x)
def f2(x):
    return 2 + 0.5*np.sqrt(x)
```

```python
x_l, x_r = 1 , 10
M,N = 10, 10
x = np.linspace(x_l, x_r, N)
y1 = f1(x)
y2 = f2(x)
y_l = np.linspace(y1[0],y2[0],M)
y_r = np.linspace(y1[-1],y2[-1],M)

plt.scatter(x,y1, s=10)
plt.scatter(x,y2, s=10)
plt.scatter([x[0]]*len(y_l), y_l, s=10)
plt.scatter([x[-1]]*len(y_r), y_r, s=10)
plt.grid()
```

канал ABCD: А левый нижний угол, а дальше против часовой стрелки

```python
def show_2D_grid(points):
    plt.figure(figsize=(20,5))
    plt.scatter(points[0], points[1], c='black', s=10)
    plt.grid(True)
```

```python
# расчётная область
N, M = 10, 5
ksi = np.arange(N)
nu = np.arange(M)
ksi_star = normalization(ksi)
nu_star = normalization(nu)
KSI_STAR, NU_STAR = np.meshgrid(ksi_star, nu_star)
show_2D_grid((KSI_STAR,NU_STAR))
```

```python
# точки на левой границе
S_AD = nu_star #Здесь должна быть сжимающая функция

# точки на правой границе
S_BC = nu_star #Здесь должна быть сжимающая функция

# значения s между поверхностями
S = S_AD + ksi_star.reshape((N,1))@(S_BC - S_AD).reshape((1,M)) #здесь ksi просто нормализация на единичный интервал


# точки на нижней границе
r_AB = ksi_star #Здесь должна быть сжимающая функция
x_AB = x_l + (x_r - x_l)*r_AB
y_AB = f1(x_AB)

# точки на верхней границе
r_DC = ksi_star #Здесь должна быть сжимающая функция 
x_DC = x_l + (x_r - x_l)*r_DC
y_DC = f2(x_DC)
```

```python
X = (1-S)*x_AB.reshape((N,1)) + S*x_DC.reshape((N,1))
Y = (1-S)*y_AB.reshape((N,1)) + S*y_DC.reshape((N,1))
```

```python
show_2D_grid((X,Y))
```

Теперь реализуем это со сжатием:

```python
class oneD_compress():
    def __init__ (self, P=1, Q=1e-10):
        """
        Это класс, реализующий одномерное сжатие. Пока он умеет сжимать только с помощью функции Робертса,
        поэтому класс содержит поля P и Q, которые являются параметрами этой функции.
        """
        self.P = P
        self.Q = Q
        return
    def RobertsCompress(self, x_star):
        """
        Сжимающая функция Робертса.
        Вход:
            x_star - np.array(n, ) - вектор нормализованных координат
        Выход:
            S - - np.array(n, ) - вектор нормализованных сжатых/растянутых координат
        """
        S = self.P*x_star + (1-self.P)*(1 - np.tanh(self.Q*(1 - x_star))/np.tanh(self.Q))
        return S
```

```python
# Создадим сжимающие функции для границ канала:
SAB = oneD_compress(P=1, Q=1e-10)
SBC = oneD_compress(P=1, Q=1e-10)
SAD = oneD_compress(P=1, Q=1e-10)
SDC = oneD_compress(P=1, Q=1e-10)
```

```python
def rect2pipe(N, M, SAB, SBC, SAD, SDC, f_down, f_up, x_l, x_r):
    """
    Функция, переводящая прямоугольную расчётную область в область внутри канала.
    Вход:
        N, M - int, int - количество ячеек по оси кси и эта в прямоугольной расчётной области
        SAB, SBC, SAD, SDC - oneD_compress() - сжимающие функции для границ канала
        f_down, f_up - function(x) - функции, задающие нижнию и верхнюю границу канала
        x_l, x_r - float, float - координаты левого и правого концов канала
    Выход:
        X - np.array(N, M) - x-координаты точек в канале, соответствующие точкам в прямоугольной расчётной области
        Y - np.array(N, M) - y-координаты точек в канале, соответствующие точкам в прямоугольной расчётной области
    """
    ksi = np.arange(N)
    nu = np.arange(M)
    ksi_star = normalization(ksi)
    nu_star = normalization(nu)
    
    
    # точки на левой границе
    S_AD = SAD.RobertsCompress(nu_star)

    # точки на правой границе
    S_BC = SBC.RobertsCompress(nu_star)

    # значения s между поверхностями
    S = S_AD + ksi_star.reshape((N,1)) @ (S_BC - S_AD).reshape((1,M))

    # точки на нижней границе
    r_AB = SAB.RobertsCompress(ksi_star)
    x_AB = x_l + (x_r - x_l)*r_AB
    y_AB = f_down(x_AB)

    # точки на верхней границе
    r_DC = SDC.RobertsCompress(ksi_star)
    x_DC = x_l + (x_r - x_l)*r_DC
    y_DC = f_up(x_DC)
    
    # почленное матричное умножение
    X = (1-S)*x_AB.reshape((N,1)) + S*x_DC.reshape((N,1)) 
    Y = (1-S)*y_AB.reshape((N,1)) + S*y_DC.reshape((N,1))
        
    return X, Y
```

```python
# напомним, как выгладит канал
x_l, x_r = 1 , 10
M,N = 10, 5
x = np.linspace(x_l, x_r, N)
y1 = f1(x)
y2 = f2(x)
y_l = np.linspace(y1[0],y2[0],M)
y_r = np.linspace(y1[-1],y2[-1],M)

plt.scatter(x,y1, s=10)
plt.scatter(x,y2, s=10)
plt.scatter([x[0]]*len(y_l), y_l, s=10)
plt.scatter([x[-1]]*len(y_r), y_r, s=10)
plt.grid()
```

```python
# равномерная сетка
SAB = oneD_compress(P=1, Q=1e-10)
SBC = oneD_compress(P=1, Q=1e-10)
SAD = oneD_compress(P=1, Q=1e-10)
SDC = oneD_compress(P=1, Q=1e-10)
N, M = 10, 5
x_l, x_r = 1, 10

X, Y = rect2pipe(N, M, SAB, SBC, SAD, SDC, f1, f2, x_l, x_r)
show_2D_grid((X, Y))
```

```python
# сгущение к началу канала
SAB = oneD_compress(P=0.1, Q=5)
SBC = oneD_compress(P=1, Q=1)
SAD = oneD_compress(P=1, Q=1)
SDC = oneD_compress(P=0.1, Q=5)
N, M = 10, 5
x_l, x_r = 1, 10

X, Y = rect2pipe(N, M, SAB, SBC, SAD, SDC, f1, f2, x_l, x_r)
show_2D_grid((X, Y))
```

```python
# сгущение к концу канала
SAB = oneD_compress(P=3, Q=1)
SBC = oneD_compress(P=1, Q=1)
SAD = oneD_compress(P=1, Q=1)
SDC = oneD_compress(P=3, Q=1)
N, M = 10, 5
x_l, x_r = 1, 10

X, Y = rect2pipe(N, M, SAB, SBC, SAD, SDC, f1, f2, x_l, x_r)
show_2D_grid((X, Y))
```
