```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
constants:
x' = ax - bxy
y' = cxy - dy
"""
a, b, c, d = 20, 5, 0.5, 2
```

```python
# right-sight function
def f(t, X):
    return np.array([a * X[0] - b * X[0] * X[1], c * X[0] * X[1] - d * X[1]])
```

```python
# runge-kutta 4
def rg4(f, t0, t1, X0, h):
    N = int((t1 - t0) / h)
    OUT = np.zeros((N, X0.size))
    OUT[0] = X0
    T = np.linspace(t0, t1, N)
    for i in range(0, N - 1):
        k1 = f(T[i], OUT[i])
        k2 = f(T[i] + h / 3, OUT[i] + k1 * h / 3)
        k3 = f(T[i] + 2 * h / 3, OUT[i] - k1 * h / 3 + k2 * h)
        k4 = f(T[i] + h, OUT[i] + k1 * h - k2 * h + k3 * h)
        OUT[i + 1] = OUT[i] + h * (k1 / 8 + 3 * k2 / 8 + 3 * k3 / 8 + k4 / 8)
    return OUT
```

```python
"""
t0 - stary time
t1 - end time
h - step
out - array with "grass" - out[:, 0] and "rabbits" - out[:, 1]
"""
t0, t1, h = 0, 10, 0.01
out = rg4(f, t0, t1, np.array([10, 2]), h)
t = np.linspace(t0, t1, int((t1 - t0) / h))
plt.plot(t, np.array(out[:,0]))
plt.plot(t, np.array(out[:,1]))
```
